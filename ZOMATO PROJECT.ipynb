## STEP 1: IMPORT NECESSARY LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## STEP2: CREATE THE  DATA FRAME

```python
df=pd.read_csv("C:/Users/vamsi/Downloads/Zomato data .csv")
```

```python
print(df.head())
```

```python
df
```

### Let's  Covert the datatype  of the " rate " column to float and remove Denominator

```python
def Rate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)
```

```python
df['rate']=df['rate'].apply(Rate)
print(df.head())
```

## THE MAJORITY OF RESTAURANTS REVEIVED RATINGS

```python
data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes': data})
plt.plot(result,c='green',marker='o')
plt.xlabel("Type of Restaurant", c="red",size=20)
plt.ylabel("votes", c="red",size=20)
```
